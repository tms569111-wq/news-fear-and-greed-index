```python
import konlpy
import time 
konlpy.__version__
import requests
import pyautogui
import openpyxl
from datetime import datetime
from bs4 import BeautifulSoup

#---네이버뉴스 크롤링---

#검색 데이터 입력
keyword="이노진"
end_num="10"
sort_num="1"
date1_num="2020.01.01"
for qqi in dates:
    date_num=qqi
    #엑셀파일
    excel_file=openpyxl.Workbook() #파일생성
    excel_file.remove_sheet(excel_file['Sheet']) #sheet 제거
    excel_sheet=excel_file.create_sheet('Results') #저장할 sheet 생성
    excel_sheet['A1']='제목'
    excel_sheet['B1']='언론사'
    excel_sheet['C1']='내용요약'
    excel_sheet['D1']='라벨링'
    excel_sheet['F1']='기사주소'

# 긍부정 단어장 생성
    cnt_neg=0 #부정 단어장
    cnt_pos=0 #긍정 단어장
    f1 = open('D:/Downloads/pos_pol_word.txt', "r", encoding="utf-8")
    pos_data=f1.read().splitlines()
    f1.close()
    f2 = open('D:/Downloads/neg_pol_word.txt', "r", encoding="utf-8")
    neg_data=f2.read().splitlines()
    f2.close()

    
    page_num=1 #페이지 초기값
    column=2 #엑셀 초기값
    for start_num in range(1,int(end_num)*10,10):
        print(f"{page_num}page 입니다")

        headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

        response=requests.get(f"https://search.naver.com/search.naver?where=news&sm=tab_pge&query={keyword}&sort={sort_num}&photo=0&field=0&pd=3&ds={date1_num}.&de={date_num}.&mynews=0&office_type=0&office_section_code=0&news_office_checked=&office_category=0&service_area=0&nso=so:dd,p:from20210101to20211211,a:all&start={start_num}", headers=headers)
        html=response.text
        soup=BeautifulSoup(html,'html.parser')
        links=soup.select(".news_tit")
        presses=soup.select(".info.press")
        contents=soup.select(".api_txt_lines") #print(links), 결과는 리스트
        column=(page_num-1)*10+2

    #링크, 기사주소
        for link in links:
            title=link.text
            url=link.attrs['href'] #href의 속성값 가져옴
        #print(title,url)
            excel_sheet[f'A{column}']=title
            excel_sheet[f'F{column}']=url
            column=column+1
   
        column=column-10  #column초기화
    
    #언론사
        for press in presses:
            press=press.text
        #print(press)
            excel_sheet[f'B{column}']=press
            column=column+1

      
        column=column-10 #column초기화
    
    #기사 본문요약 
        for content in contents:
            content = content.text
            for i in neg_data:
                if i in content:
                    cnt_neg = cnt_neg + 1
            for j in pos_data:
                if j in content:
                    cnt_pos = cnt_pos + 1
            if cnt_pos >= cnt_neg:
                excel_sheet[f'D{column}'] = "1"
            elif cnt_pos < cnt_neg:
                excel_sheet[f'D{column}'] = "0"
            excel_sheet[f'C{column}'] = content
            cnt_pos = 0
            cnt_neg = 0
            column = column + 1


        page_num=page_num+1 #페이지 수 보정

        time.sleep(2) 
#크롤링 시각 제목으로 엑셀 저장    
    file_name=datetime.now().strftime("%Y-%m-%d %H-%M-%S")
    excel_file.save('C:/news/'+file_name+'.xlsx') #저장경로 및 파일 제목, 반드시 경로 변경 후 사용 바람
    import konlpy

    konlpy.__version__
#삼성 데이터 엑셀 파일 주피터로 train, text 데이터
    import pandas as pd
    import openpyxl
    def read_data(filename):
        data = pd.read_excel(filename, usecols=[1,2,3])
        data = data[1:]
        return data

    train_data = read_data('C:/news/2023-11-15 16-41-24.xlsx').values.tolist()
    test_data = read_data('C:/news/'+file_name+'.xlsx').values.tolist()

    from konlpy.tag import Okt

    okt = Okt()
    import json
    import os
    from pprint import pprint

    def tokenize(doc):
        try:
            return ['/'.join(t) for t in okt.pos(doc, norm=True, stem=True)]
        except Exception as e:
            print(f"Exception in tokenize: {e}")
            return []

    train_docs = [(tokenize(str(row[1])), row[2]) for row in train_data]
    train_docs = [tuple(i) for i in train_docs]
    test_docs=[(tokenize(str(row[1])),row[2]) for row in test_data]
    test_docs = [tuple(i) for i in test_docs]

    tokens = [t for d in train_docs for t in d[0]]

    import nltk
    text = nltk.Text(tokens, name='NMSC')


# 중복을 제외한 토큰의 개수
    shape=int(len(set(text.tokens)))
          

# 시간이 꽤 걸립니다! 시간을 절약하고 싶으면 most_common의 매개변수를 줄여보세요.
    selected_words = [f[0] for f in text.vocab().most_common(10000)]

    def term_frequency(doc):
        return [doc.count(word) for word in selected_words]

    train_x = [term_frequency(d) for d, _ in train_docs]
    test_x = [term_frequency(d) for d, _ in test_docs]
    train_y = [c for _, c in train_docs]
    test_y = [c for _, c in test_docs]
    import numpy as np

    x_train = np.asarray(train_x).astype('float32')
    x_test = np.asarray(test_x).astype('float32')

    y_train = np.asarray(train_y).astype('float32')
    y_test = np.asarray(test_y).astype('float32')
    from tensorflow.keras import models
    from tensorflow.keras import layers
    from tensorflow.keras import optimizers
    from tensorflow.keras import losses
    from tensorflow.keras import metrics

# 실제 데이터의 특성 수에 맞게 변경
    input_shape = shape  # 예시로 592로 설정, 실제 데이터의 특성 수에 맞게 변경해야 함

    model = models.Sequential()
    model.add(layers.Dense(64, activation='relu', input_shape=(input_shape,)))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))

    model.compile(optimizer=optimizers.RMSprop(learning_rate=0.001),
                 loss=losses.binary_crossentropy,
                 metrics=[metrics.binary_accuracy])

    model.fit(x_train, y_train, epochs=10, batch_size=512)
    results = model.evaluate(x_test, y_test)
    LS=0
    ct=0
    def predict_pos_neg(review):
        token = tokenize(review)
        tf = term_frequency(token)
        data = np.expand_dims(np.asarray(tf).astype('float32'), axis=0)
        score = float(model.predict(data))
        global LS
        LS=LS+score
        global ct
        ct=ct+1
    last_eval=read_data('C:/news/'+file_name+'.xlsx').values.tolist()
    for i in last_eval:
        predict_pos_neg(i[1])
    everydata.append(float(LS/ct))
print("EVERYTHING is DONEEEEE")
```
